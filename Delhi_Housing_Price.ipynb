```python
grouped_data=df.groupby('Bedrooms')['Bedrooms'].count()
plt.bar(grouped_data.index,grouped_data.values)
plt.xlabel('Bedrooms')
plt.ylabel('Count')
plt.title('Count of Bedrooms')
plt.xticks(grouped_data.index)
plt.plot()
```

```python
grouped_data=df.groupby('Bathrooms')['Bathrooms'].count()
plt.bar(grouped_data.index,grouped_data.values)
plt.xlabel('Bathrooms')
plt.ylabel('Count')
plt.title('Count of Bathrooms')
plt.xticks(grouped_data.index)
plt.plot()
```

```python
grouped_data=df.groupby('Balcony')['Balcony'].count()
plt.bar(grouped_data.index,grouped_data.values)
plt.xlabel('Balcony')
plt.ylabel('Count')
plt.title('Count of Balcony')
plt.xticks(grouped_data.index)
plt.show()
```

```python
grouped_data=df.groupby('Status')['Status'].count()
plt.pie(grouped_data,labels=grouped_data.index,autopct='%1.2f%%')
plt.title('Status')
plt.legend()
plt.show()
```

```python
grouped_data=df.groupby('neworold')['neworold'].count()
plt.pie(grouped_data,labels=grouped_data.index,autopct='%1.2f%%')
plt.title('Neworold')
plt.legend()
plt.show()
```

```python
grouped_data=df.groupby('type_of_building')['type_of_building'].count()
plt.pie(grouped_data,labels=grouped_data.index,autopct='%1.2f%%')
plt.title('Type of Buildings')
plt.legend()
plt.show()
```

```python
plt.hist(df['price'],bins='auto')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Price Distribution')
plt.show()
```

```python
plt.figure(figsize=(11,16))
pivot_table=df.pivot_table(values='price',index='area',columns='Bedrooms',aggfunc='mean')
sns.heatmap(pivot_table, cmap='coolwarm')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Area')
plt.title('Heatmap - Price by Area and Bedrooms')
plt.show()
```

```python
import folium
from folium.plugins import MarkerCluster
```

```python
map_center = [28.6139, 77.2090]  
map_zoom = 10  
map_delhi = folium.Map(location=map_center, zoom_start=map_zoom)
marker_cluster = MarkerCluster().add_to(map_delhi)
folium.Marker(map_center, popup = "New Delhi").add_to(map_delhi)
for index, row in df.iterrows():
    lat = row['latitude']
    lon = row['longitude']
    folium.Marker([lat, lon]).add_to(marker_cluster)
map_delhi
```

```python
map_center = [28.6139, 77.2090]  
map_zoom = 10  
map_delhi = folium.Map(location=map_center, zoom_start=map_zoom)
folium.Marker(map_center, popup = "New Delhi").add_to(map_delhi)
Lat_Lon = df[['latitude','longitude']].to_numpy()
folium.plugins.HeatMap(Lat_Lon,radius=10).add_to(map_delhi)
display(map_delhi)
```

```python
df['Weight']=0.5
map_center = [28.6139, 77.2090]  
map_zoom = 10  
Lat_Lon = df[['latitude','longitude','Weight']].to_numpy()
map_delhi = folium.Map(map_center, zoom_start=10)
folium.Marker(map_center, popup = "New Delhi").add_to(map_delhi)
folium.plugins.HeatMap(Lat_Lon, radius=15).add_to(map_delhi)

display(map_delhi)
```
